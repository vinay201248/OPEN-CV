```python
import numpy as np 
import cv2
```

`creating a Trackbar for some Problem Statements:`

### Task-1

```python
def fun(x):
    print(x)
    
```

```python
cv2.namedWindow('check')
cv2.createTrackbar('Basic' , 'check' , 0, 255 , fun)

while True:
    black_image = np.zeros([512,512,3] , dtype = np.float64)
    d = cv2.getTrackbarPos('Basic' , 'check')
    d = int(d)
    text = str(d)
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(black_image , text , (150,150) , font , 1 , (255,255,255)  ,1)
    cv2.imshow('check' , black_image)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np 

def fun(x):
    print(x)
    
cv2.namedWindow('teja')

cv2.createTrackbar('l_h' , 'teja', 0,255 , fun)
cv2.createTrackbar('l_s','teja', 0,255 , fun)
cv2.createTrackbar('l_v','teja', 0,255 , fun)


cv2.createTrackbar('h_h','teja', 255,255 , fun)
cv2.createTrackbar('h_s','teja', 255,255 , fun)
cv2.createTrackbar('h_v','teja', 255,255 , fun)


while True:
    color_image = cv2.imread('./../images/tulips.jpg',1)
    hsv_image = cv2.cvtColor(color_image , cv2.COLOR_BGR2HSV)

    
    l_h = int(cv2.getTrackbarPos('l_h','teja'))
    l_s = int(cv2.getTrackbarPos('l_s','teja'))
    l_v = int(cv2.getTrackbarPos('l_v','teja'))
    
    h_h = int(cv2.getTrackbarPos('h_h','teja'))
    h_s = int(cv2.getTrackbarPos('h_s','teja'))
    h_v = int(cv2.getTrackbarPos('h_v','teja'))
    
    lower_range = np.array([l_h , l_s , l_v])
    upper_range = np.array([h_h , h_s , h_v])
    
    mask = cv2.inRange(hsv_image , lower_range , upper_range)
    final = cv2.bitwise_and(color_image , color_image , mask = mask)

    
    cv2.imshow('teja' , final)
    cv2.imshow('mask',mask)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```

`For Bluring an Image we are using following 4 ways :`

- Homogenius methods (Filter2d) (uses mean of neighbourhood pixels)
- Bilateral Method (Blur) (uses average same as homogenious)
- Gaussian filter (Gaussian)  (uses high pixel in the middle and low pixels in the edges)
- Median (just uses median of the Neighbourhood)

- for making blur we need to use low intensity values inside kernels 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
color_image = cv2.imread('./../images/lena_color_512.tif' ,1)


cv2.imshow('image' ,  color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Filter2d:

kernal = np.array([[1,1,1] , [1,1,1] ,[1,1,1]] , dtype = np.float32) / 9.0


blur_res = cv2.filter2D(color_image , -1 , kernal)
cv2.imshow('blur' , blur_res)
cv2.imshow('color' , color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
kernal = np.ones((7,7) , np.float64) / 49.0
blur_res = cv2.filter2D(color_image , -1 , kernal)
cv2.imshow('blur' , blur_res)
cv2.imshow('color' , color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# blur
kernal = np.ones((5,5) , np.float64) / 25.0
blur_res = cv2.filter2D(color_image , -1 , kernal)
blur_res_1 = cv2.blur(color_image , (5,5))
cv2.imshow('blur' , blur_res_1)
cv2.imshow('filter' , blur_res)
cv2.imshow('color' , color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# gaussian filter 

blur_res_1 = cv2.GaussianBlur(color_image , (7,7), 0)
cv2.imshow('gaussian' , blur_res_1)

cv2.imshow('color' , color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# median 

blur_res_1 = cv2.medianBlur(color_image , 9)
cv2.imshow('gaussian' , blur_res_1)

cv2.imshow('color' , color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Image Sharpen 

```python
kernal = np.array([[0,-1,0] , [-1,5,-1] ,[0,-1,0]] , dtype = np.float32)


sharp = cv2.filter2D(color_image , -1 , kernal)
cv2.imshow('s' , sharp)
cv2.imshow('color' , color_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Edge Detection 


```python
color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)
kernal = np.array([[0,1,0] , [1,-4,1] ,[0,1,0]] , dtype = np.float32)
edge = cv2.filter2D(gray , -1 , kernal)
cv2.imshow('edge' , edge)
cv2.imshow('color' , color_image)
cv2.imshow('gray',gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

## Canny Edge Detection 

- 1.Noise Reduction 
- 2.gradient Calculation 
- 3.Non-max Supression
- 4.Double threhold
- 5.Edge Tracking by Hystresis

```python
color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)
edge = cv2.Canny(gray , 100 , 100)
cv2.imshow('edge' , edge)
cv2.imshow('color' , color_image)
cv2.imshow('gray',gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

- selecting threshold values for canny edge detection using Trackbar:

```python
def fun(x):
    print(x)
    
cv2.namedWindow('open')

cv2.createTrackbar('t1' , 'open' , 0 , 255 , fun)
cv2.createTrackbar('t2' , 'open' , 0 , 255 , fun)

while True:
    color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
    gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)
    
    value1 = cv2.getTrackbarPos('t1' , 'open')
    value2 = cv2.getTrackbarPos('t2' , 'open')
    
    canny_edge = cv2.Canny(gray , value1 , value2)
    
    cv2.imshow('open' , canny_edge)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```

### Simple Thresholding concepts 

```python
# Binary means < 150 converts to black and > 180 converts to white 

color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)

a , th = cv2.threshold(gray , 150 , 180 , cv2.THRESH_BINARY)
cv2.imshow('orginal' , gray)
cv2.imshow('binary' , th)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Binary inverse  < 150 converts to white and > 180 converts to black 

color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)

a , th = cv2.threshold(gray , 150 , 180 , cv2.THRESH_BINARY_INV)
cv2.imshow('orginal' , gray)
cv2.imshow('binary' , th)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# threshold_to_zero <  127 converts to 0 and > 127 remains constant 

color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)

a , th = cv2.threshold(gray , 127 , 255 , cv2.THRESH_TOZERO)
cv2.imshow('orginal' , gray)
cv2.imshow('binary' , th)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# threshold_to_zero <  127 remains constant and > 127 converts to zero

color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)

a , th = cv2.threshold(gray , 127 , 255 , cv2.THRESH_TOZERO_INV)
cv2.imshow('orginal' , gray)
cv2.imshow('binary' , th)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Threshold_trunc means at the value 127 remains same and after 127 remains as 127 color only 

color_image = cv2.imread('./../images/lena_color_512.tif' ,1)
gray = cv2.cvtColor(color_image , cv2.COLOR_BGR2GRAY)

a , th = cv2.threshold(gray , 127 , 255 , cv2.THRESH_TRUNC)
cv2.imshow('orginal' , gray)
cv2.imshow('binary' , th)
cv2.waitKey()
cv2.destroyAllWindows()
```

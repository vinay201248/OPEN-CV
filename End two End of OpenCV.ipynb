```python
# live time recording 
# saving video and saving each frame:
# videos:

# more about get method : https://stackoverflow.com/questions/39953263/get-video-dimension-in-python-opencv


import datetime

video = cv2.VideoCapture('./../images/test_video.mp4')  # 0-webcam
# print(video.read())  # 2 outcomes : True or Flase , Pixels 

print(video.get(3))  # width 
print(video.get(4))  # height

count = 0
while True:
    res , frame = video.read()
    if res == True:
        text = str(datetime.datetime.now())
        font = cv2.FONT_HERSHEY_COMPLEX
        cv2.putText(frame , text , (100 , 50) , font , 1 , (0,0,255) , 1)
        cv2.imwrite('./../Frames/'+str(count)+'.png' , frame)
        cv2.imshow('win' , frame)
        count+=1
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break
    else:
        break
        
video.release()
cv2.destroyAllWindows()
```

```python

import datetime

video = cv2.VideoCapture('./../images/test_video.mp4')  # 0-webcam
# print(video.read())  # 2 outcomes : True or Flase , Pixels 

print(video.get(3))  # width 
print(video.get(4))  # height

width = video.get(3)
height = video.get(4)



forcc = cv2.VideoWriter_fourcc(*'XVID')  # saving the video instead if individual frames:
output = cv2.VideoWriter('./../images/output.avi' , forcc , 20.0 , (int(width) , int(height)))

while True:
    res , frame = video.read()
    if res == True:
        text = str(datetime.datetime.now())
        font = cv2.FONT_HERSHEY_COMPLEX
        cv2.putText(frame , text , (100 , 50) , font , 1 , (0,0,255) , 1)
        output.write(frame)
        cv2.imshow('win' , frame)
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break
    else:
        break
        
video.release()
cv2.destroyAllWindows()
```

```python
# adding two images:
# when ur adding two images , two images shape should be same:

image_1 = cv2.imread('./../images/lena_color_256.tif' , 1)
image_2 = cv2.imread('./../images/peppers_color.tif' , 1)
print(image_1.shape)
print(image_2.shape)

res = cv2.resize(image_2 , (256 , 256))
c = cv2.add(image_1 , res)

cv2.imshow('image_1' , image_1)
cv2.imshow('image_2' , res)

cv2.imshow('addresult' , c)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# add weightage


image_1 = cv2.imread('./../images/lena_color_256.tif' , 1)
image_2 = cv2.imread('./../images/peppers_color.tif' , 1)
print(image_1.shape)
print(image_2.shape)

res = cv2.resize(image_2 , (256 , 256))
c = cv2.addWeighted(image_1 , 0.3 , res , 0.7 , 1)

cv2.imshow('image_1' , image_1)
cv2.imshow('image_2' , res)

cv2.imshow('addresult' , c)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Bitwise operators :

# AND , OR , NOT , XOR :

image1 = np.zeros([512 , 512 , 3] , np.uint8)
image2 = np.zeros([512 , 512 , 3] , np.uint8)
cv2.rectangle(image1 , (256 , 0) , (512,512), (255,255,255),-1)
cv2.rectangle(image2 , (150 , 0) , (350,200), (255,255,255),-1)

# Logical and 
logical_and = cv2.bitwise_and(image1 , image2)

# logical or 

logical_or = cv2.bitwise_or(image1 , image2)

# logical xor:

logical_xor = cv2.bitwise_xor(image1 , image2)

logical_not = cv2.bitwise_not(image1)

cv2.imshow('image_1' , image1)
#cv2.imshow('image_2' , image2)
# cv2.imshow('and' , logical_and)
# cv2.imshow('or' , logical_or)
# cv2.imshow('xor' , logical_xor)
cv2.imshow('not' , logical_not)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# HSV [Hue Saturation Value:]

# because HSV means the output will be in human visualization manner:

image = cv2.imread('./../images/tulips.jpg')

hsv = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)

cv2.imwrite('./../images/hsv.png' , hsv)
cv2.imshow('BGR' , image)
cv2.imshow('HSV' , hsv)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# selecting only red color using HSV

image = cv2.imread('./../images/tulips.jpg')
hsv = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)

lower_limit = np.array([20, 100, 100])
upper_limit = np.array([30, 255, 255])

# ranges can be selected from some internet:

mask = cv2.inRange(hsv , lower_limit , upper_limit)

result = cv2.bitwise_and(image , image , mask = mask)

cv2.imshow('original' , image)

cv2.imshow('mask' , mask)
cv2.imshow('HSV' , hsv)

cv2.imshow('res' , result)
cv2.waitKey()
cv2.destroyAllWindows()
```

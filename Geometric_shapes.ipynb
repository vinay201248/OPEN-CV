```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./images/mandril_color.tif' , 1)
plt.imshow(image[: ,: ,::-1])
```

### Rectangle

```python
cv2.rectangle(image , (100 , 100) , (200,200),(0,0,255) , 3)
plt.imshow(image[:,:,::-1])
```

```python
cv2.rectangle(image , (100 , 100) , (200,200),(0,0,255) , -1)
plt.imshow(image[:,:,::-1])
```

```python
cv2.imshow('image' , image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Line

```python
plt.imshow(image[:,:,::-1])
```

```python
cv2.line(image , (50,80) , (120,140) , (255,0,0) , 3)
plt.imshow(image[:,:,::-1])
```

```python
cv2.arrowedLine(image , (150,80) , (180,140) , (255,0,0) , 3)
plt.imshow(image[:,:,::-1])
```

### Circle

```python
### Putting All together 

black_image = np.zeros((512 , 512 , 3) , dtype = np.float64)
cv2.imshow('black' , black_image)

# Rectange:
cv2.rectangle(black_image , (100 , 100) , (200,200),(0,0,255) , 3)
# Line 
cv2.line(black_image , (50,80) , (120,140) , (255,0,0) , 3)
# Arrowed_line
cv2.arrowedLine(black_image , (150,80) , (180,140) , (255,0,0) , 3)
# circle
cv2.circle(black_image , (350,350) , 35 , (0,255,0),-1)
# text
text = 'Data Scientist'
font = cv2.FONT_HERSHEY_COMPLEX

cv2.putText(black_image , text , (100,200) ,font, 1 , (255,0,0) , 2)

cv2.imshow('black_1' , black_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

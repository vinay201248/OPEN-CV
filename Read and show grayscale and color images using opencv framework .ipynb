#### Checking each channel in color images which means RGB 


```python
image = plt.imread('./images/mandril_color.tif')
plt.subplot(2,2,1)
plt.title('original_image')
plt.xticks([])
plt.imshow(image)

plt.subplot(2,2,2)
plt.title('R_channel')
plt.xticks([])
plt.imshow(image[: , : , 0] , cmap='gray')

plt.subplot(2,2,3)
plt.title('G_channel')
plt.xticks([])
plt.imshow(image[: , : , 1] , cmap='gray')


plt.subplot(2,2,4)
plt.title('B_channel')
plt.xticks([])
plt.imshow(image[: , : , 2] , cmap='gray')
```

### checking channels with opencv

```python
image = cv2.imread('./images/mandril_color.tif' , 1)  # reading with opencv 

plt.subplot(2,2,1)
plt.title('original_image')
plt.xticks([])
plt.imshow(image[: , : , ::-1])

plt.subplot(2,2,2)
plt.title('B_channel')
plt.xticks([])
plt.imshow(image[: , : , 0] , cmap='gray')

plt.subplot(2,2,3)
plt.title('G_channel')
plt.xticks([])
plt.imshow(image[: , : , 1] , cmap='gray')


plt.subplot(2,2,4)
plt.title('R_channel')
plt.xticks([])
plt.imshow(image[: , : , 2] , cmap='gray')
```

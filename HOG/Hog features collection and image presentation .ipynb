```python
from skimage import data , feature
import matplotlib.pyplot as plt
```

```python

img = plt.imread('./xYjKoa.jpg')
plt.imshow(img)
```

```python
vectors , hog_image = feature.hog(img , orientations=9 ,pixels_per_cell=(8, 8), cells_per_block=(2, 2),block_norm='L2',
    visualize=True,multichannel=True , channel_axis=2)
```

```python
plt.figure(figsize = (13,13))
plt.subplot(1,2,1)
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(hog_image)
plt.xticks([])
plt.yticks([])
plt.title('Hog_image')
plt.show()
```
